## 1.환경준비

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 타이타닉 데이터
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

## 2. seaborn 차트

### histogram

```python
sns.histplot(data=titanic, x='Age', bins= 16)
plt.show()
```

```python
sns.histplot(data=titanic, x='Age', bins=16, hue='Survived')
plt.show()
```

### densityplot

```python
sns.kdeplot(data=titanic, x='Age')
plt.show()
```

```python
sns.kdeplot(data=titanic, x='Age', hue='Survived')
plt.show()
```

```python
sns.kdeplot(data=titanic, x='Age', hue='Survived', common_norm=False)
plt.show()
```

### boxplot

```python
sns.boxplot(data=titanic, y='Age')
plt.show()
```

```python
sns.boxplot(data=titanic, x='Survived', y='Age')
plt.show()
```
